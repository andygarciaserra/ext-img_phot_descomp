```python
#Imports:
import os
import numpy as np
from astropy.io import fits
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats

#DIRECTORIES:
datadir = 'data/'
figdir = 'figures/'
```

```python
#Importing data:
data = np.loadtxt(datadir+'MCG-01-10-015_i.txt',skiprows=2)
r = data[:,0]
lgr = np.log10(r)

    #Intensity:
I = data[:,1]
lgI = np.log10(I)
plt.figure()
plt.plot(r,lgI)
plt.plot([0,140],[3.3,0],'r')
plt.xlim([0,40])
plt.ylim([2,3])
plt.axvline(6, color='r', linestyle='-')
plt.savefig(figdir+'r_I.png',bbox_inches='tight',dpi=300)
r = r.flatten().tolist()
pos = r.index(6)
Io = I[pos]
print(Io)

    #Ellipticity:
el = data[:,3]
elo = el[pos]
print(elo)
plt.figure()
plt.plot(r,el)
plt.axhline(y=0.69, color='r', linestyle='-')
plt.savefig(figdir+'r_el.png',bbox_inches='tight',dpi=300)
    
    #PA:
PA = data[:,5]
lgPA = np.log10(PA)
PAo = PA[pos]
print(PAo)
plt.figure()
plt.plot(r,PA)
plt.axhline(y=-7.2, color='r', linestyle='-')
plt.savefig(figdir+'r_pa.png',bbox_inches='tight',dpi=300)
```
